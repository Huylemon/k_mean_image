```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = mpimg.imread("input_btl.png")
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
X = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
# Cu lnh img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
# c s dng  thay i hnh dng ca mng img thnh mt mng hai chiu.
for K in [2, 5, 10, 15, 20]:
    # mt vng lp c thc hin qua danh sch cc gi tr K: 2, 5, 10, 15, v 20.
    kmeans = KMeans(n_clusters=K).fit(X)
    # thut ton K-means c p dng vo d liu X bng cch s dng KMeans(n_clusters=K).fit(X).
    # iu ny to ra mt m hnh K-means vi s lng nhm l K v thc hin vic gn nhn cho cc
    # im d liu trong X bng cch s dng kmeans.predict(X).
    label = kmeans.predict(X)

    # mt mng img4 c to ra c kch thc ging vi X, nhng ton b cc
    # gi tr c khi to bng 0 (np.zeros_like(X)).
    img4 = np.zeros_like(X)
    # cc im d liu trong img4 c nhn tng ng vi k c thay th bng gi tr
    # trung tm ca nhm  (kmeans.cluster_centers_[k]).
    for k in range(K):
        img4[label == k] = kmeans.cluster_centers_[k]
    # reshape and display output image
    # Dng code img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    # chuyn i img4 t mt mng hai chiu thnh mt mng c kch thc ban u ca hnh nh img.
    img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
```
