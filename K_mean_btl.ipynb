```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = mpimg.imread("input_btl.png")
plt.imshow(img)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
X = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
# Câu lệnh img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
# được sử dụng để thay đổi hình dạng của mảng img thành một mảng hai chiều.
for K in [2, 5, 10, 15, 20]:
    # một vòng lặp được thực hiện qua danh sách các giá trị K: 2, 5, 10, 15, và 20.
    kmeans = KMeans(n_clusters=K).fit(X)
    # thuật toán K-means được áp dụng vào dữ liệu X bằng cách sử dụng KMeans(n_clusters=K).fit(X).
    # Điều này tạo ra một mô hình K-means với số lượng nhóm là K và thực hiện việc gán nhãn cho các
    # điểm dữ liệu trong X bằng cách sử dụng kmeans.predict(X).
    label = kmeans.predict(X)

    # một mảng img4 được tạo ra có kích thước giống với X, nhưng toàn bộ các
    # giá trị được khởi tạo bằng 0 (np.zeros_like(X)).
    img4 = np.zeros_like(X)
    # các điểm dữ liệu trong img4 có nhãn tương ứng với k được thay thế bằng giá trị
    # trung tâm của nhóm đó (kmeans.cluster_centers_[k]).
    for k in range(K):
        img4[label == k] = kmeans.cluster_centers_[k]
    # reshape and display output image
    # Dòng code img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    # chuyển đổi img4 từ một mảng hai chiều thành một mảng có kích thước ban đầu của hình ảnh img.
    img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
```
